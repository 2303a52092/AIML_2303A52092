```python
car_train = train_data.drop('binaryClass', axis=1)
car_labels = train_data[['binaryClass']]
```

```python
from sklearn.preprocessing import OrdinalEncoder

from sklearn.pipeline import Pipeline

pipe = Pipeline([('ord_enc', OrdinalEncoder())])

car_train_prepared = pipe.fit_transform(car_train)
```

```python
from sklearn.preprocessing import LabelEncoder

label_enc = LabelEncoder()

car_labels_prepared = label_enc.fit_transform(car_labels)
```

```python
car_train_prepared.shape, car_labels_prepared.shape
```

```python
from sklearn.tree import DecisionTreeClassifier

tree_clf = DecisionTreeClassifier()
tree_clf.set_params(class_weight='balanced', criterion='entropy')

tree_clf.fit(car_train_prepared, car_labels_prepared)
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(20,15))
plot_tree(tree_clf, feature_names=car_train.columns,class_names=['N','P']);
```

```python
from sklearn.metrics import accuracy_score, classification_report,\
confusion_matrix

preds = tree_clf.predict(car_train_prepared)
print(f"Accuracy:{accuracy_score(car_labels_prepared, preds)}")
print(f"ClassReport:{classification_report(car_labels_prepared,preds)}")
print(f"Confusion Matrix:{confusion_matrix(car_labels_prepared,preds)}")
```

```python
from sklearn.model_selection import GridSearchCV

params_grid = {'max_leaf_nodes': list(range(0, 10)),
               'min_samples_split': [0,1,2, 3, 4],
               'min_samples_leaf': [0,1,2, 3, 4],
               'max_depth':[0,1,2,3,4,5],
               'max_features':[0,1,2,3,4,5],
              'max_leaf_nodes':[0,1,2,3,4,5]}

#refit is true by default. The best estimator is trained on the whole dataset

grid_search = GridSearchCV(DecisionTreeClassifier(random_state=42, class_weight='balanced'), params_grid, verbose=1, cv=3, refit=True)

grid_search.fit(car_train_prepared, car_labels_prepared)
```

```python
grid_search.best_params_
```

```python
grid_search.best_estimator_
```

```python
tree_best = grid_search.best_estimator_
```

```python
from sklearn.metrics import accuracy_score

# Assuming 'tree_best' is your trained model, 'car_train_prepared' is your feature data, and 'car_labels_prepared' are your true labels
accuracy = accuracy_score(car_labels_prepared, tree_best.predict(car_train_prepared))
print(accuracy)
```

```python
from sklearn.metrics import confusion_matrix

def conf_matrix(X, model, y_true):
    """
    Generates and prints the confusion matrix.

    Args:
        X: The feature data.
        model: The trained model.
        y_true: The true labels.
    """
    y_pred = model.predict(X)
    cm = confusion_matrix(y_true, y_pred)
    print(cm)

# Example usage:
conf_matrix(car_train_prepared, tree_best, car_labels_prepared)
```

```python
from sklearn.metrics import classification_report  # Import the correct function

# Assuming 'tree_best' is your trained model, 'car_train_prepared' is your feature data, and 'car_labels_prepared' are your true labels
report = classification_report(car_labels_prepared, tree_best.predict(car_train_prepared))
print(report)
```

```python
car_test = test_data.drop('binaryClass', axis=1)
car_test_labels = test_data['binaryClass']
```

```python
from sklearn.metrics import accuracy_score  # Import the accuracy_score function

def accuracy(X, model, y):
    """Calculates the accuracy of the model."""
    predictions = model.predict(X)
    acc = accuracy_score(y, predictions) # Use accuracy_score function
    return acc

# Now you can call the accuracy function
accuracy(car_test_prepared, tree_best, car_test_labels_prepared)
```

```python
from sklearn.metrics import classification_report # Import the classification_report function

def class_report(X, model, y):
    """Generates a classification report."""
    predictions = model.predict(X)
    report = classification_report(y, predictions) # Use classification_report function
    print(report)  # Print the report

# Now you can call the class_report function
class_report(car_test_prepared, tree_best, car_test_labels_prepared)
```
